```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x = [1, 5, 4]
y = [23, 45, 89]
x2 = [3, 7, 3]
y2 = [5, 8, 9]

plt.plot(x, y, label = "First", linewidth = 5)
plt.plot(x2, y2, label = "Second")
plt.legend() # For labelling .legend() has to be run
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x = [1, 5, 4]
y = [23, 45, 89]
x2 = [3, 7, 3]
y2 = [5, 8, 9]

plt.scatter(x, y, label = "First", linewidth = 5) # scatter uses points to show graph
plt.plot(x2, y2, label = "Second")
plt.legend() # For labelling .legend() has to be run
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x = [1, 5, 4]
y = [23, 45, 89]
x2 = [3, 7, 3]
y2 = [5, 8, 9]

plt.hist(x, y, label = "First", linewidth = 5) # Histogram uses points to show graph
plt.plot(x2, y2, label = "Second")
plt.legend() # For labelling .legend() has to be run
plt.show()
```
